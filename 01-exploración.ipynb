##EXPLORACION SENCILLA DE LOS DATOS
A continuación se realizará una exploración a los datos de las pruebas saber pro 2025-2.

```python
# Lista completa de programas académicos con número de estudiantes
programas_completos = df["E_PRGM_ACADEMICO"].value_counts()

print("\nLista completa de programas académicos con numero de estudiantes:")
print(programas_completos)

# Número de programas distintos
print(f"\nNúmero total de programas académicos: {df['E_PRGM_ACADEMICO'].nunique()}")

# Número total de estudiantes
print(f"Número total de estudiantes: {df['E_PRGM_ACADEMICO'].count()}")
```

```python
# Top 10 programas académicos con más estudiantes
top10_programas = df["E_PRGM_ACADEMICO"].value_counts().head(10)

print("\nTop 10 programas académicos con más estudiantes:")
print(top10_programas)
```

```python
# Número de estudiantes por departamento
estudiantes_departamento = df["E_PRGM_DEPARTAMENTO"].value_counts()

print("\nDistribución de estudiantes por departamento:")
print(estudiantes_departamento)

# Número total de departamentos distintos
print(f"\nNúmero total de departamentos: {df['E_PRGM_DEPARTAMENTO'].nunique()}")
```

```python
#Agrupamos por programa académico y rango de matrícula
matricula_por_programa = (
    df.groupby(["E_PRGM_ACADEMICO", "E_VALORMATRICULAUNIVERSIDAD"])
      .size()
      .reset_index(name="Conteo")   # convierte índice en columnas
)

# Mostrar el resultado (solo primeras 20 filas)
print(matricula_por_programa.head(20))
```

```python
#Conocer si hay algún estudiante privado de la libertad
# Crear la tabla cruzada
tabla = pd.crosstab(df['E_PRGM_ACADEMICO'], df['E_PRIVADO_LIBERTAD'])

# Renombrar las columnas para mayor claridad
tabla = tabla.rename(columns={0: 'No_Privados', 1: 'Privados'})

# Agregar una columna con el total
tabla['Total'] = tabla.sum(axis=1)

# Mostrar la tabla final
print(tabla)
```

```python
# Clasificación del Rendimiento global de los estudiantes por programa
# Tabla de distribución de rendimiento por programa académico
tabla_rendimiento = pd.crosstab(
    df['E_PRGM_ACADEMICO'],
    df['RENDIMIENTO_GLOBAL']
)

# Ordenar las columnas
orden = ['bajo', 'medio-bajo', 'medio', 'medio-alto', 'alto']
tabla_rendimiento = tabla_rendimiento.reindex(columns=orden, fill_value=0)

# Agregar columna con el total de estudiantes por programa
tabla_rendimiento['Total'] = tabla_rendimiento.sum(axis=1)

# Ordenar por total (de mayor a menor)
tabla_rendimiento = tabla_rendimiento.sort_values(by='Total', ascending=False)

# Mostrar los primeros 20 programas
print(tabla_rendimiento.head(20))
```
